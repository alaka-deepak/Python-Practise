### Python Inheritance
<li>Inheritance allows us to define a class that inherits all the methods and properties from another class.
<li>Parent class is the class being inherited from, also called base class.
<li>Child class is the class that inherits from another class, also called derived class.

<li>Note: The __init__() function is called automatically every time the class is being used to create a new object.








#### Python also has a super() function that will make the child class inherit all the methods and properties from its parent:

```python
class person:
    def __init__(self,fname,lname):
        self.firstname=fname
        self.lastname=lname

    def printname(self):
        print(self.firstname,self.lastname)

class student(person):
    def __init__(self,fname,lname,year):
        super().__init__(fname,lname)
        self.graduationyear=year

    def printname(self):
        print(self.firstname,self.lastname,self.graduationyear)

x=student("mike","ef",1934)
x.printname()
```

#### Polymorphism

<li> programming it refers to methods/functions/operators with the same name that can be executed on many objects or classes.

```python
class job1():
    def __init__(self,course):
        self.course_offered=course
    def display_course(self):
        print("DATA SCIENCE")

class job2:
    def __init__(self,course):
        self.course_offered=course
    def display_course(self):
        print("AI ENGINEER")

j1=job1("DS")
j2=job2("AI")

# for x in (j1,j2):
#     x.move()
j1.display_course()
j2.display_course()
```

### Python Inner class

<li> An inner class is a class defined inside another class. 
<li> The inner class can access the properties and methods of the outer class.



```python
# Inner class called by outer class
class Outer:
    def __init__(self):
        self.name="Outer Class"
    
    
    class Inner:
        def __init__(self,outer):
            self.outer=outer
            # self.name="Inner class"

        def display_msg(self):
            print("Hai Good morning")
            print(f"Outer class can see here: {self.outer.name}")

outer=Outer()
# print(outer.name)
inner=outer.Inner(outer)
inner.display_msg()
# print(inner.display_msg())

# Messages are print from right to left

```
