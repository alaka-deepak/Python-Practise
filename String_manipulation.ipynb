### Slicing

```python
s_sl="Hello World"
print(s_sl[2:5])

# Slice form start

print(s_sl[:5]) #5 not inluded

# Slice to the end
print(s_sl[2:])
# Negative indexes to start from end -2 not included
print(s_sl[-5:-2])
```

### F-Strings

```python
# String Format
age=36 
txt="My name is noha" #+age is incorrect
# To specify a string as an f-string, simply put an f in front of the string literal, and add curly brackets {} as placeholders for variables and other operations.
print(f"{txt} and age is {age} years old")
```

```python
# A placeholder can include a modifier to format the value.
# A modifier is included by adding a colon : followed by a legal formatting type, like .2f which means fixed point number with 2 decimals:

print(f"The Age is {age:.2f} years old")
print(f"The price is {20*59} dollars")
```

```python
# Escape Characters
print(f"{txt} \"Lucky\" form North of England")
```

```python
#Count how many times each character appears in a string
#Input: "banana" → Output: {'b': 1, 'a': 3, 'n': 2}
txt="I am Noha and age is 30 years old"
freq={}

for i in txt:
    freq[i]=freq.get(i,0)+1

print(freq)

```

### Problems 

```python
# Remove all duplicate characters
#Input: "programming" → Output: "progamin"

print(txt)
uni=[]
for i in txt:
    if i in uni:
        continue
    else:
        uni.append(i)

print(uni)
```

```python
#Find the most frequent character in a string
#Input: "success" → Output: 's'

txt="malayalam"

fre={}
for i in txt:
    fre[i]=fre.get(i,0)+1

most_freq=max(fre,key=fre.get)

print(most_freq)
```

```python

# Check if two strings are anagrams
# Input: "listen", "silent" → Output: True

str1="hello"
str2="edsfs"

if len(str1 )!= len(str2):
    print("Not")
else:
    fre1={}
    fre2={}
    for i in str1:
        fre1[i]=fre1.get(i,0)+1
    for j in str2:
        fre2[j]=fre2.get(j,0)+1
    print(fre1==fre2)
```

```python
# Remove all punctuation from a string
# Input: "Hello, world!" → Output: "Hello world"

txt="(:>{mayaAppu"
import string
result=""
for i in txt:
    if i not in string.punctuation:
        result+=i

print("Wihout punctuation",result)
```

```python
# Find the first non-repeating character in a string
# Input: "aabbcde" → Output: 'c'
nonre={}
for i in txt:
    nonre[i]=nonre.get(i,0)+1

minfre=min(nonre.values())

# Collect all the characters

least_charact=[]
least_charact=[ch for ch in nonre if nonre[ch]==minfre]
    # ch here is taking keys -characters
print(least_charact)
```

#### Count vowels and consonants

```python
txt=input("Enter the string")
vowels="aeiouAEIOUS"
v=sum(1 for i in txt if i in vowels)
c=sum(1 for i in txt if i.isalpha() and i not in vowels)
print(f"Name is :{txt} Vowels:{v} and consonants:{c}")
```
